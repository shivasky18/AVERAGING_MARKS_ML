```python

data3=pd.read_csv('/SLIPTEST3.csv')
# print(data.columns)
datac=data3[['Roll Number','Total score']]
print(data3[['Roll Number','Total score','Name of the Student']])
```

```python
# Convert 'Roll Number' to a common data type (e.g., string) in all DataFrames
dataa['Roll Number'] = dataa['Roll Number'].astype(str)
datab['Roll Number'] = datab['Roll Number'].astype(str)
datac['Roll Number'] = datac['Roll Number'].astype(str)

# Now perform the merge operations
# Change on=['Roll Number,Name of the Student'] to on=['Roll Number', 'Name of the Student']
merged_data1_2 = pd.merge(dataa, datab, on='Roll Number', how='outer', suffixes=('_test1', '_test2'))
data= pd.merge(merged_data1_2, datac, on='Roll Number', how='outer', suffixes=('_test1_test2', '_test3'))
print(data)
```

```python
# import pandas as pd

# # Load your dataset
# data = pd.read_csv('your_file.csv')  # Replace with the actual file name

# Function to convert 'x.xx / 10' string to a float
def convert_to_float(score_str):
    if pd.isna(score_str):
        return None
    return float(score_str.split(' / ')[0])

# Apply the function to each score column
data['Score_test1'] = data['Total score_test1'].apply(convert_to_float)
data['Score_test2'] = data['Total score_test2'].apply(convert_to_float)
data['Score_test3'] = data['Total score'].apply(convert_to_float)

# Define a function to calculate the best two scores average or single score average with others
def best_or_single_avg(scores):
    # Remove None values (missing scores)
    scores = [score for score in scores if score is not None]

    # If there are fewer than 1 valid score, return None
    if len(scores) < 1:
        return None

    # If there is only 1 valid score, average it with the other two if possible
    if len(scores) == 1:
        # Find the missing scores
        missing_scores = [score for score in [scores[0], None, None] if score is None]
        # If two scores are missing, return the single score
        if len(missing_scores) == 2:
            return scores[0]
        else:
            # Average the single score with available other scores
            valid_scores = [scores[0]] + [score for score in scores[1:] if score is not None]
            return sum(valid_scores) / len(valid_scores)

    # If there are two valid scores, return the average of those two
    if len(scores) == 2:
        return sum(scores) / 2

    # If there are three valid scores, return the average of the best two
    scores.sort(reverse=True)
    return sum(scores[:2]) / 2

# Apply the function across rows to calculate the best two averages
data['Best_Two_Average'] = data[['Score_test1', 'Score_test2', 'Score_test3']].apply(best_or_single_avg, axis=1)

# Print the result
print(data[['Roll Number', 'Best_Two_Average']])

# Optionally, save the result to a new CSV file
# data.to_csv('best_two_average_result.csv', index=False)
```

```python
import pandas as pd
import math

# Assuming 'data' contains the merged data with 'Roll Number', 'Total score_test1', 'Total score_test2'
# Step 1: Convert 'NaN' values in the test scores to 0 for students who have only one test score
data['Total score_test1'] = data['Total score_test1'].apply(lambda x: float(str(x).split(' / ')[0]) if pd.notna(x) else 0)
data['Total score_test2'] = data['Total score_test2'].apply(lambda x: float(str(x).split(' / ')[0]) if pd.notna(x) else 0)

# Step 2: Calculate the average score for each student (with respect to two tests, even if one test score is missing)
data['Average Score'] = (data['Total score_test1'] + data['Total score_test2']) / 2

# Step 3: Apply math.ceil() to round up the average score to the nearest integer
data['Average Score'] = data['Average Score'].apply(lambda x: math.ceil(x))

# Step 4: Divide the average score by 2
data['Average Score Divided by 2'] = data['Average Score'] / 2

# Step 5: Apply math.ceil() to round up the result of the division
data['Final Average Score'] = data['Average Score Divided by 2'].apply(lambda x: math.ceil(x))

# Step 6: Print the updated dataframe with roll numbers and their final scores
# print(data[['Roll Number', 'Total score_test1', 'Total score_test2', 'Final Average Score']])
data1=data[['Roll Number','Final Average Score']]
print(data1)
data1.to_csv('final_data.csv', index=False)
# print(data)
```
